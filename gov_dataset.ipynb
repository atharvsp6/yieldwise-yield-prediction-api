```python
import requests
import time
import os

API_KEY = "579b464db66ec23bdd000001858b953682a348a845b40b532b0a03ea"
RESOURCE_ID = "35be999b-0208-4354-b557-f6ca9a5355de"
FILE_NAME = "gov_crop_data.csv"

BASE_URL = f"https://api.data.gov.in/resource/{RESOURCE_ID}"
LIMIT = 500
MAX_RETRIES = 5
DELAY = 1


# ---------- Find last offset from file ----------
def get_last_offset():
    if not os.path.exists(FILE_NAME):
        return 0
    with open(FILE_NAME, "rb") as f:
        rows = sum(1 for _ in f) - 1  # minus header
        return max(rows, 0)


offset = get_last_offset()
print("▶ Resuming from offset:", offset)


# ---------- Download loop ----------
while True:
    params = {
        "api-key": API_KEY,
        "format": "csv",
        "limit": LIMIT,
        "offset": offset
    }

    success = False

    for attempt in range(MAX_RETRIES):
        try:
            r = requests.get(BASE_URL, params=params, timeout=30)
            r.raise_for_status()
            success = True
            break
        except Exception as e:
            print(f"⚠️ Offset {offset} failed (attempt {attempt+1}/{MAX_RETRIES}) → {e}")
            time.sleep(3)

    if not success:
        print("❌ Too many failures. Moving to next chunk.")
        offset += LIMIT
        continue

    content = r.content

    if len(content) < 100:
        print("✅ All records downloaded.")
        break

    file_exists = os.path.exists(FILE_NAME)

    with open(FILE_NAME, "ab") as f:
        if not file_exists:
            f.write(content)
        else:
            f.write(b"\n".join(content.splitlines()[1:]))

    offset += LIMIT
    print("✔ Saved till offset:", offset)

    time.sleep(DELAY)

print("🎉 Final CSV ready:", FILE_NAME)
```
